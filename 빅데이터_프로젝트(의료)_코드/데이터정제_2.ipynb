```python
import pandas as pd
```

```python
df = pd.read_csv("./df_dn.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
#환자 ID, 혈액형
df.drop(['환자ID','혈액형'], axis=1, inplace=True)
```

```python
# 림프구 
df = df[df['Large Lymphocyte'] >= 7]
```

```python
# ODI 0 채우기
df.ODI.fillna(0, inplace=True)
```

```python
# 가족력 drop
df.dropna(subset=['가족력'],inplace = True)
```

```python
df = df[df["우울증여부"] != 2]
# 우울증여부의 이상치 제거(2)
```

```python
df['입원일자'] = pd.to_datetime(df['입원일자'], format='%Y-%m-%d')
df['퇴원일자'] = pd.to_datetime(df['퇴원일자'], format='%Y-%m-%d')
df['최근입원기간'] = df['퇴원일자'] - df['입원일자']
df['누적입원기간'] = df['입원기간'] * 30
df["최근입원기간"] = df["최근입원기간"].astype(str)
df["최근입원기간"] = df["최근입원기간"].str.replace(' days','')
df["최근입원기간"] = df["최근입원기간"].astype(int)
df['입원기간이상치'] = (df['최근입원기간'] <= df['누적입원기간'])
df = df[df['입원기간이상치'] == True]
df.drop(['입원기간이상치'], axis=1, inplace=True)
```

```python
#직업
df["직업"].fillna("기타", inplace = True)
```

```python
#헤모글로빈 성별 기준 평균값

헤모글로빈수치_1 = df[df['성별'] == 1]['헤모글로빈수치'].mean()
헤모글로빈수치_2 = df[df['성별'] == 2]['헤모글로빈수치'].mean()

df.헤모글로빈수치.fillna(헤모글로빈수치_1, inplace=True)
```

```python
# 통증기간
df['통증기간(월)'].fillna(df['통증기간(월)'].mean(skipna= True), inplace = True)
```

```python
# 수술시간
df['수술시간'].fillna(df['수술시간'].mean(skipna= True), inplace = True)
```

```python
#수술기법 
df["수술기법"].fillna("기타", inplace = True)
```

```python
df = df[df['후방디스크높이(mm)'] < 100] 
```

```python
# 지방축적도
df.dropna(subset=['지방축적도'],inplace = True)
df = df[df['지방축적도'] < 12000] 
```

```python
# PI
df.dropna(subset=['PI'],inplace = True)
df = df[df['PI'] < 400] 
```

```python
# PT
df.dropna(subset=['PT'],inplace = True)
```

```python
df.dropna(subset=['Seg Angle(raw)'],inplace = True)
df = df[df['Seg Angle(raw)'] < 150] 
```

```python
df.dropna(subset=['디스크단면적'],inplace = True)
```

```python
df = df[df['디스크위치'] < 7]
```
