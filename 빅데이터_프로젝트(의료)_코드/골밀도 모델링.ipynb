```python
df = pd.get_dummies(image, columns = ['척추이동척도'])
```

```python
df.dropna(inplace = True)
```

```python
for item in df.columns:
    print(item, end = '+')
```

```python
df.columns
```

```python
df.rename(columns={'Seg Angle(raw)' : 'Seg_Angle(raw)'}, inplace= True)
df.rename(columns={'Vaccum disc' : 'Vaccum_disc'}, inplace= True)
df.rename(columns={'Modic change' : 'Modic_change'}, inplace= True)

df.rename(columns={'척추이동척도_Extremely down' : '척추이동척도_Extremely_down'}, inplace= True)
df.rename(columns={'척추이동척도_Extremely up' : '척추이동척도_Extremely_up'}, inplace= True)
df.rename(columns={'척추이동척도_Extremely up' : '척추이동척도_Extremely_up'}, inplace= True)
df.rename(columns={'MF + ES' : 'MF_ES'}, inplace= True)

df.rename(columns={'전방디스크높이(mm)' : '전방디스크높이'}, inplace= True)
df.rename(columns={'후방디스크높이(mm)' : '후방디스크높이'}, inplace= True)
df.rename(columns={'Seg_Angle(raw)' : 'Seg_Angle'}, inplace= True)

```

```python
reg_model = smf.ols(formula='골밀도 ~전방디스크높이+후방디스크높이+지방축적도+Instability+MF_ES+Modic_change+PI+PT+Seg_Angle+Vaccum_disc+디스크단면적+디스크위치+척추전방위증+척추이동척도_Down+척추이동척도_Extremely_down+척추이동척도_Extremely_up+척추이동척도_Middle+척추이동척도_Up', data = df)
reg_result = reg_model.fit()
print(reg_result.summary())
```
